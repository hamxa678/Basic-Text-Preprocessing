##Applying Text Processing techniques on description column

1. Converting it to lowercase

```python
movies_df['description'] = movies_df['description'].str.lower()
movies_df.head()
```

2. Removing HTML tags

```python
import re

def remove_html(text):
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)

movies_df['description'] = movies_df['description'].apply(remove_html)
movies_df.head()
```

3. Removing URLs

```python
def remove_url(text):
  pattern = re.compile('https?://\S+|www\.\S+')
  return pattern.sub(r'', text)

movies_df['description'] = movies_df['description'].apply(remove_url)
```

```python
movies_df.head()
```

4. Removing Punctuations

```python
import string

def replace_punctuation_with_space(text):
    punctuation_chars = string.punctuation
    translator = str.maketrans(punctuation_chars, ' ' * len(punctuation_chars))
    return text.translate(translator)

# Assuming movies_df is already defined and has a 'description' column
movies_df['description'] = movies_df['description'].apply(replace_punctuation_with_space)
movies_df.head()
```

5. Spelling Correction

```python
from textblob import TextBlob

def correct_spelling(text):
    blob = TextBlob(text)
    return str(blob.correct())

movies_df['description'] = movies_df['description'].apply(correct_spelling)
movies_df.head()
```

6. Removing Stop words

```python
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords
stopwords.words('english')

def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))
    words = text.split()
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return ' '.join(filtered_words)

movies_df['description'] = movies_df['description'].apply(remove_stopwords)
movies_df.head()
```

```python
stemmed_df = movies_df.copy()
stemmed_df.head()
```

7. Stemming

```python
import nltk
nltk.download('punkt')

from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

def nltk_stem(text):
    stemmer = PorterStemmer()
    tokens = word_tokenize(text)
    stemmed_tokens = [stemmer.stem(token) for token in tokens]
    return ' '.join(stemmed_tokens)

stemmed_df['description'] = stemmed_df['description'].apply(nltk_stem)
stemmed_df.head()
```

8. Tokenization

```python
import spacy

nlp = spacy.load('en_core_web_sm')

def tokenize_text(text):
    doc = nlp(text)
    return [token.text for token in doc]

stemmed_df['description'] = stemmed_df['description'].apply(tokenize_text)
stemmed_df.head()
```

9. Lemmetization

```python
import nltk
nltk.download('wordnet')

from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

def lemmatize_text(text):
    tokens = nltk.word_tokenize(text)
    lemmatized_tokens = [lemmatizer.lemmatize(token, pos='v') for token in tokens]
    return ' '.join(lemmatized_tokens)

movies_df['description'] = movies_df['description'].apply(lemmatize_text)
movies_df.head()
```
